```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os

sys.path.append(os.path.abspath('../src'))

from data_loader import load_data
```

```python
data = load_data("processed.csv", "Processed", False)
```

```python
data['Exited'].value_counts().sort_index() 
```

```python
counts = data['Exited'].value_counts().sort_index() 
plt.bar(counts.index.astype(str), counts.values, color=(sns.color_palette("Set2")))
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Counts of customers that did not exit (0) and the ones that exited (1)')
plt.show()
```
