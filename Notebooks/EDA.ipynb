```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os

sys.path.append(os.path.abspath('../src'))

from data_loader import load_data
```

```python
data = load_data("processed.csv", "Processed", False)
```

```python
data['Exited'].value_counts().sort_index() 
```

```python
counts = data['Exited'].value_counts().sort_index() 
plt.bar(counts.index.astype(str), counts.values, color=(sns.color_palette("Set2")))
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Counts of customers that did not exit (0) and the ones that exited (1)')
plt.show()
```

```python
country_values = data[data.Exited == 1].Geography.value_counts(ascending=True)
```

```python
x = [0, 1, 2]
plt.barh(country_values.index[x], country_values.iloc[x], color=sns.color_palette("mako_r"))
plt.title("Counts of exited customers per each country")
plt.xlim(0, 900)
plt.show()
```
